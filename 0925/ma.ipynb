```python
import pandas as pd
import numpy as np

from datetime import datetime,date
from tqsdk import TqApi,TqAuth,TqBacktest,TargetPosTask

api=TqApi(web_gui=True,backtest=TqBacktest(start_dt=date(2014,1,1),end_dt=date(2024,9,1)),auth=TqAuth("18049858285","18049858285wk"))
```

```python
week_klines = api.get_kline_serial("SHFE.rb2501", 7*86400, data_length=15)
day_klines=api.get_kline_serial("SHFE.rb2501", 86400)

target_pos = TargetPosTask(api, "SHFE.rb2501")
```

```python
while True:
    api.wait_update()
    if api.is_changing(day_klines):
        pre_ma5=sum(week_klines.close.iloc[-6:-1]) / 5
        pre_ma10 = sum(week_klines.close.iloc[-11:-1]) / 10
        ma5 = sum(week_klines.close.iloc[-5:]) / 5
        ma10 = sum(week_klines.close.iloc[-10:]) / 10
        print("最新价", day_klines.close.iloc[-1], "MA5", ma5)
        print("最新价", day_klines.close.iloc[-1], "MA10", ma10)
        if (pre_ma5<pre_ma10) and (ma5>ma10):
            print("最新价大于MA: 目标多头5手")
            # 设置目标持仓为多头5手
            target_pos.set_target_volume(5)
        elif (pre_ma5>pre_ma10) and (ma5<ma10):
            print("最新价小于MA: 目标空仓")
            # 设置目标持仓为空仓
            target_pos.set_target_volume(0)
```
